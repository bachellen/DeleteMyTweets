```python
#!/usr/bin/env python

import argparse
import csv
import sys
import time
import os
import twitter
from dateutil.parser import parse

def delete(api):
    with open("output_got.csv") as file:
        count = 0

        for row in csv.DictReader(file):
            tweet_id = int(row["id"])
            try:
                print ("Deleting tweet", tweet_id)

                api.DestroyStatus(tweet_id)
                count += 1
                time.sleep(0.5)

            except:
                print ("Exception")

    print ("Number of deleted tweets:", count)

# def error(msg, exit_code=1):
#     sys.stderr.write("Error: %s\n" % msg)
#     exit(exit_code)

def main():
    CONSUMER_KEY = ''
    CONSUMER_SECRET = ''
    ACCESS_TOKEN  = ''
    ACCESS_SECRET = ''
    api = twitter.Api(CONSUMER_KEY,
                     CONSUMER_SECRET,
                      ACCESS_TOKEN,
                      ACCESS_SECRET)
    delete(api)


if __name__ == "__main__":
    main()
```
